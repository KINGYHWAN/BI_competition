```python
# 숫자 내 comma 제거

def remove_rest(x):
    return float(x.replace("," , ""))
```

```python
# 데이터 내 필요한 기간만 사용

industrial_production_indicator = industrial_production_indicator[3:141]
industrial_production_indicator = industrial_production_indicator.reset_index(drop = True)
GDP_data = GDP_data[2:48]
GDP_data = GDP_data.reset_index(drop = True)
GDP_data['국내총생산 및 경제성장률'] = GDP_data['국내총생산 및 경제성장률'].apply(remove_rest)
```

```python
# 분기 GDP --> 월별 GDP

temp_result = []

for i in GDP_data.index:

    a = float(industrial_production_indicator['전산업생산지수(원지수)_분기,월'][3*i])

    b = float(industrial_production_indicator['전산업생산지수(원지수)_분기,월'][3*i+1])

    c = float(industrial_production_indicator['전산업생산지수(원지수)_분기,월'][3*i+2])

    sum  = a + b + c

    mon_result1 = (a / sum) * float(GDP_data['국내총생산 및 경제성장률'][i])
    mon_result2 = (b / sum) * GDP_data['국내총생산 및 경제성장률'][i]
    mon_result3 = (c / sum) * GDP_data['국내총생산 및 경제성장률'][i]

#    print(mon_result1, mon_result2, mon_result3)

    temp_result.append(mon_result1)
    temp_result.append(mon_result2)
    temp_result.append(mon_result3)


print(len(temp_result))
print(temp_result)
```
